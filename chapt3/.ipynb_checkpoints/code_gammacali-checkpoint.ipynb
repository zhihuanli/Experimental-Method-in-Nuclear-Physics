````c++
double fun(double *x, double *par)//+
{
   return  par[0]*TMath::Exp(-0.5*( (x[0]-par[1])/par[2])  * ( (x[0]-par[1])/par[2]) ) + par[3]+x[0]*par[4];
}
````

````c++
TF1 *ffit = new TF1("ffit",fun,0,2500,5);// 0  2500   5 
ffit->SetParameter(0,1.0e6);//par[0]-
ffit->SetParameter(1,332.0);//par[1]-
ffit->SetParameter(2,1.0);//par[2]-sigma
ffit->SetParameter(3,1.0e3);//par[3]-
ffit->SetParameter(4,0.01);//par[4]-

h_gammaraw->GetXaxis()->SetRangeUser(0,2500);//
h_gammaraw->Fit("ffit","","",328,338);
h_gammaraw->Draw();
h_gammaraw->GetXaxis()->SetRangeUser(325,340);
c1->Draw();
````

````c++
    double p0 = 0;//
    double p1 = 1;
    double p2 = 0;
    TH1D *h1 = new TH1D("h1","",10000,0,2500);//h1gamma
    TRandom3 *r = new TRandom3(0);
    int Nbins = h0->GetXaxis()->GetNbins(); //h0gamma
    for(int i=0; i<Nbins; i++)
    {
        
        Long64_t eN = h0->GetBinContent(i);// GetBinContent()bin
        Double_t e = h0->GetBinLowEdge(i);// GetBinLowEdge()binx
        for(Long64_t j=0; j<eN; j++)
        {
            Double_t ea = e+r->Rndm()*0.2;// 
            ea=p2*ea*ea+p1*ea+p0;//p0,p1,p2
            h1->Fill(ea);
        }
    }
h1->Draw();
c1->Draw();
````
